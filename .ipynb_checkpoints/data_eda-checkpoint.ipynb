```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('mlb_data_cleaning.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
#displaying the range of players ages
sns.set_style('whitegrid')
df['Age'].hist(bins=30)
plt.xlabel('Age')
```

```python
#checking if age and Games are corrolated
sns.jointplot(x='Age',y='G',data=df)
```

```python
#checking if age has something to do with hitting
sns.jointplot(x='Age',y='H',data=df, kind = 'reg')
```

```python
#does age and batting average have something in commmon?
sns.jointplot(x='Age',y='BA',data=df,color='red',kind='kde');
plt.savefig("AgetoBA.png")
```

```python
#hits to batting average
sns.jointplot(x='H',y='BA',data=df,color='green', kind = 'reg')
plt.savefig("hitstoBA.png")
```

```python
#plate appearance to batting average
sns.jointplot(x='PA',y='BA',data=df,color='blue', kind = 'reg')
```

```python
#at bats to batting average
sns.jointplot(x='AB',y='BA',data=df, kind = 'reg')
plt.savefig("At_batstoBA.png")
```

```python
#heat map of all the factors to see corrolation
cmap = sns.diverging_palette(220, 10, as_cmap = True)
sns.heatmap(df[['Age','H','BA', 'G', 'PA', 'HR', 'RBI', 'SO','BB']].corr(), vmax = .3, center = 0, cmap=cmap,
           square = True, linewidths = .5, cbar_kws = {'shrink':.5})
plt.savefig("Heatmap.png")
```

```python
#batting average to strikeouts
sns.jointplot(x='BA',y='SO',data=df)
```
