```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
sns.set_style('whitegrid')
df['Age'].hist(bins=30)
plt.xlabel('Age')
```

```python
sns.jointplot(x='Age',y='G',data=df)
```

```python
sns.jointplot(x='Age',y='H',data=df, kind = 'reg')
```

```python
sns.jointplot(x='Age',y='BA',data=df,color='red',kind='kde');
```

```python
sns.jointplot(x='H',y='BA',data=df,color='green', kind = 'reg')
```

```python
cmap = sns.diverging_palette(220, 10, as_cmap = True)
sns.heatmap(df[['Age','H','BA', 'G', 'PA', 'HR', 'RBI', 'SO','BB']].corr(), vmax = .3, center = 0, cmap=cmap,
           square = True, linewidths = .5, cbar_kws = {'shrink':.5})
```
